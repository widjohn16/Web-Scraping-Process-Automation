```python
#Importa as bibliotecas para processar os arquivos baixados..
import os
folder = r'c:\SOLAR\\'
os.listdir(folder)
os.rename(folder + 'marcacoes_0002.txt' , folder + 'REL012.txt')
os.rename(folder + 'marcacoes_0003.txt' , folder + 'REL014.txt')
os.rename(folder + 'marcacoes_0004.txt' , folder + 'REL013.txt')
os.rename(folder + 'marcacoes_0001.txt' , folder + 'REL015.txt')


import os

folder = r"C:\SOLAR\\"  # substitua pelo caminho correto
arquivos = os.listdir(folder)

for file in arquivos:
    if file.endswith(".txt"):
        if "REL012" in file:
            destino = os.path.join(folder, "REL_012")
        elif "REL013" in file:
            destino = os.path.join(folder, "REL_013")
        elif "REL014" in file:
            destino = os.path.join(folder, "REL_014")
        elif "REL015" in file:
            destino = os.path.join(folder, "REL_015")
        else:
            continue

        if not os.path.exists(destino):
            os.makedirs(destino)
        os.rename(os.path.join(folder, file), os.path.join(destino, file))
    
```
